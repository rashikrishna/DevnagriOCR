## This program creates a csv file for our image where each column depicts the pixel value and the 'label' column depicts the actual character label.

```python
from PIL import Image
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#JUST CHECKING
imgdir=os.path.join(os.getcwd(),'Dataset/vowels/a/008_01.jpg')
img = Image.open(imgdir)
```

```python
#JUST CHECKING
a1 = np.asarray(img)
print(a1.shape)
a1=a1/255
a1=1-a1
plt.imshow(a1)
```

```python
'''Converting images to pandas dataframe where each column represents a pixel
Since each image is 28X28, hence total number of columns are 784'''
i=1
df=pd.DataFrame()
imgdir=os.path.join(os.getcwd(),'Dataset')
for char_type in os.listdir(imgdir):
    char_type_dir=os.path.join(imgdir,char_type)
    for char_file in os.listdir(char_type_dir):
        char_file_dir=os.path.join(char_type_dir,char_file)
        for img_file in os.listdir(char_file_dir):
            finalimgdir=os.path.join(char_file_dir,img_file)
            img = Image.open(finalimgdir)
            a2=np.array(img)
            a2=a2.reshape(1,784)
            df3=pd.DataFrame(a2)
            df3['labels']=char_file
            df=df.append(df3)
df
```

```python
#Saving the data frame as a csv file
df.to_csv('file1.csv')
```
